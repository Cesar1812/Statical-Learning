### Tratamiento de Outliers

```python
Columnas =data_de_interes.columns
```

```python
def detectOutlierLimits(dataset, col):

    IQR = dataset[col].quantile(0.75) - dataset[col].quantile(0.25)

    LI = dataset[col].quantile(0.25) - (IQR*1.75)

    LS = dataset[col].quantile(0.75) + (IQR*1.75)

 

    return LI, LS
```

```python
LI, LS = detectOutlierLimits(data_de_interes, Columnas)
LI, LS
```

```python
import pandas as pd
from sklearn.preprocessing import RobustScaler


# Seleccionar las columnas num√©ricas que se utilizar√°n para el modelo
numeric_cols = ['Customer_Age', 'Avg_Utilization_Ratio', 'Credit_Limit']

# Crear un objeto RobustScaler para escalar los datos
scaler = RobustScaler()

# Escalar los datos num√©ricos
data[numeric_cols] = scaler.fit_transform(data[numeric_cols])

# Calcular el rango intercuartil (IQR) para cada columna num√©rica
Q1 = data[numeric_cols].quantile(0.25)
Q3 = data[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Filtrar los valores at√≠picos utilizando el m√©todo de los valores extremos (IQR)
outliers_removed = data[~((data[numeric_cols] < (Q1 - 1.5 * IQR)) |(data[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
from scipy import stats

lambdas = []
nombre = []
df_new = pd.DataFrame()

for x in data_noout.columns:
    df, lamb = stats.boxcox(data_noout[x])
    lambdas.append(lamb)
    nombre.append(x)
    df_new[x] = df
```

```python
from sklearn.cluster import KMeans
```

```python
inercia = []
ks = []

for k in range(1,60):
    modelo_iterado = KMeans(n_clusters = k, random_state = 5)
    modelo_iterado.fit(data_bc)
    inercia.append(modelo_iterado.inertia_)
    ks.append(k)
```

```python
codo = {
    'inercia': inercia,
    'k': ks
}

codo = pd.DataFrame(codo)
```

```python
sns.scatterplot(data=codo,x='k',y='inercia',markers=True)
sns.lineplot(data=codo, x='k', y='inercia',)
```

```python
modelo = KMeans(n_clusters=3,random_state = 10)
modelo.fit(data_bc)
```
